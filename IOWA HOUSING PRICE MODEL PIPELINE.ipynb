```python
#Capturing File Path
iowa_data_full = pd.read_csv('/home/braincreator/iowa/train.csv', index_col= 'Id')
iowa_data_test_full = pd.read_csv('/home/braincreator/iowa/test.csv', index_col='Id')
```

```python
#Remove rows with missing target, separate target from predictions
iowa_data_full.dropna(axis=0, subset=['SalePrice'], inplace = True)
y = iowa_data_full.SalePrice
iowa_data_full.drop(['SalePrice'], axis=1, inplace = True)
```

```python
#Creating Train-Test Split
X_train_full, X_valid_full, y_train, y_valid = train_test_split(iowa_data_full, y, train_size = 0.8, test_size = 0.2,
                                                               random_state = 0)
```

```python
#Selecting categorical columns with relatively low cardinality
categorical_lowcard_cols = [cname for cname in X_train_full.columns if 
                   X_train_full [cname].nunique()< 10 and
                   X_train_full[cname].dtype =='object']

#Selecting numerical columns
numerical_cols = [cname for cname in X_train_full.columns if
                  X_train_full[cname]. dtype in ['int64', 'float64']]
```

```python
#Selected columns are now kept
my_cols = categorical_lowcard_cols + numerical_cols
X_train = X_train_full[my_cols].copy()
X_valid = X_valid_full[my_cols].copy()
X_test = iowa_data_test_full[my_cols].copy()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_absolute_error
```

```python
#Preprocessing for numerical data
numerical_transformer = SimpleImputer(strategy = 'constant')

#Preprocessing for categorical data
categorical_transformer = Pipeline(steps = [('imputer', SimpleImputer(strategy='most_frequent')),
                                           ('onehot', OneHotEncoder(handle_unknown = 'ignore'))])
```

```python
#Bundling preprocessing for numrical and categorical data
preprocessor = ColumnTransformer(transformers = [
    ('num', numerical_transformer, numerical_cols),
    ('cat', categorical_transformer, categorical_lowcard_cols)])
```

```python
from sklearn.ensemble import RandomForestRegressor as RFR

#Fit with Pipeline model
my_pipeline = Pipeline(steps = [('preprocessor', preprocessor),
                                   ('model', RFR(n_estimators = 100, criterion = 'mae', random_state=0))])
my_pipeline.fit(X_train, y_train)
preds = my_pipeline.predict(X_valid)
mae = mean_absolute_error(y_valid, preds)

print ('MAE:', mae)
```

```python
print (my_pipeline.predict (X_valid.head(1460)))
```

```python
from sklearn.model_selection import cross_val_score as cvs

scores = -1 * cvs(my_pipeline, iowa_data_full, y,
                  cv=5,
                  scoring ='neg_mean_absolute_error')
                
print("Average MAE score:", scores.mean())
```
