```python
#The model I will use is a Decision Tree Regressor to specified parameters of a random state of 1
#The steps I will now follow are to: Define, Fit, Create Predictions, Validate, Evaluate, and Redesign - if need be.

from sklearn.tree import DecisionTreeRegressor as DTR
```

```python
#To validate the quality of my model, I will use mean absolute error (MAE) to gain average of prediction errors within it
#To do this, I will split data into training and validation data to gain 'Out-of-sample' scores on error

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error as MAE
```

```python
train_Z, val_Z, train_y, val_y = train_test_split (Z, y, random_state = 0)
```

```python
iowa_model.fit(train_Z, train_y)
```

```python
#Calculating for MAE

val_predictions = iowa_model.predict(val_Z)
print (MAE(val_y, val_predictions))
```
